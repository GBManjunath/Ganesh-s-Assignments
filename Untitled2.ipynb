Q1. What are the characteristics of tuples? Is tuple immutable?
Characteristics of tuples:

Ordered: Tuples maintain the order of elements. The order in which elements are inserted is the order in which they are accessed.
Immutable: Once a tuple is created, its elements cannot be changed, added, or removed.
Heterogeneous: A tuple can store different data types (e.g., integers, strings, floats).
Indexed: Each element in a tuple has an index, starting from 0.
Allow duplicate values: Tuples can contain duplicate elements.
Is a tuple immutable? Yes, tuples are immutable. Once created, the values inside a tuple cannot be altered. You cannot add, remove, or modify elements in a tuple after its creation.

Q2. What are the two tuple methods in Python? Give an example of each method. Give a reason why tuples have only two in-built methods as compared to Lists.
Two tuple methods:

count(): Returns the number of occurrences of a specified value in the tuple.

python
Copy code
my_tuple = (1, 2, 3, 1, 1)
print(my_tuple.count(1))  # Output: 3
index(): Returns the index of the first occurrence of a specified value in the tuple.

python
Copy code
my_tuple = (1, 2, 3, 4)

Reason for only two in-built methods: Tuples are immutable, meaning you cannot change their content. Therefore, there are fewer methods for manipulation (such as adding, removing, or changing elements). Lists, on the other hand, are mutable, so they require more methods to modify their contents.

Q4. Explain the difference between the union() and update() methods for a set. Give an example of each method.
union() method:

Returns a new set containing all the unique elements from both sets. It does not modify the original sets.
Syntax: set1.union(set2)
Example:

python
Copy code
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1.union(set2)
print(result)  # Output: {1, 2, 3, 4, 5}
update() method:

Modifies the original set by adding elements from another set (or iterable). It does not return a new set.
Syntax: set1.update(set2)
Example:

python
Copy code
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1.update(set2)
print(set1)  # Output: {1, 2, 3, 4, 5}
Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.
A dictionary in Python is an unordered collection of key-value pairs, where each key must be unique. The values can be of any data type and can be repeated.

Example:

python
Copy code
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
Ordered or Unordered?

In Python 3.7 and later, dictionaries are ordered, meaning the insertion order of key-value pairs is preserved.
In earlier versions of Python (before 3.7), dictionaries were unordered.

Q7. Using setdefault() method, create a key named topics in the given dictionary and also add the value of the key as this list ['Python', 'Machine Learning', 'Deep Learning'].
Example:

python
Copy code
dict1 = {'language': 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
# Output: {'language': 'Python', 'course': 'Data Science Masters', 'topics': ['Python', 'Machine Learning', 'Deep Learning']}
The setdefault() method ensures that the key 'topics' is added with the specified list as its value if it doesn't already exist in the dictionary.

Q8. What are the three view objects in dictionaries? Use the three in-built methods in Python to display these three view objects for the given dictionary.
The three view objects in dictionaries are:

keys(): Returns a view object that displays a list of all the keys in the dictionary.
values(): Returns a view object that displays a list of all the values in the dictionary.
items(): Returns a view object that displays a list of tuples, each containing a key-value pair.
Example:


dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

# Displaying the keys
print(dict1.keys())  # Output: dict_keys(['Sport', 'Teams'])

# Displaying the values
print(dict1.values())  # Output: dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])

# Displaying the items
print(dict1.items())  # Output: dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])
