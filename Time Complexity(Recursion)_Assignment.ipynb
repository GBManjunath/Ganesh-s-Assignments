```python
'''Find time complexity of below code blocks :


Problem 1 :
def quicksort(arr):
if len(arr) <= 1:
return arr
pivot = arr[len(arr) // 2]
left = [x for x in arr if x < pivot]
middle = [x for x in arr if x == pivot]
right = [x for x in arr if x > pivot]
return quicksort(left) + middle + quicksort(right)

Answer 1: The time complexity of quicksort is O(n log n) in the average case. 
This is because at each level of recursion, the algorithm partitions the array into two halves, performing O(n) operations. With log n levels of recursion, the overall time complexity is O(n log n).


Problem 2 :
def nested_loop_example(matrix):
rows, cols = len(matrix), len(matrix[0])
total = 0
for i in range(rows):
for j in range(cols):
total += matrix[i][j]
return total

Answer 2: The time complexity of the nested loop example is O(rows * cols). The function iterates through each element in a matrix with 'rows' rows and 'cols' columns, resulting in a total of rows * cols iterations.

Problem 3 :
def example_function(arr):
result = 0
for element in arr:
result += element
return result

Answer 3: The time complexity of example_function is O(n), where n is the length of the array. The function iterates through each element in the array once.

Problem 4 :
def longest_increasing_subsequence(nums):
n = len(nums)
lis = [1] * n
for i in range(1, n):
for j in range(0, i):
if nums[i] > nums[j] and lis[i] < lis[j] + 1:
lis[i] = lis[j] + 1
return max(lis)


The time complexity of the longest_increasing_subsequence function is O(n^2), where n is the length of the input list. This is due to the nested loops that compare elements and update the longest increasing subsequence.

Problem 5 :
def mysterious_function(arr):
n = len(arr)
result = 0
for i in range(n):
for j in range(i, n):
result += arr[i] * arr[j]
return result

Answer 5: The time complexity of mysterious_function is O(n^2), where n is the length of the array. The function has two nested loops that iterate through each element in the array and accumulate the product of pairs.


Solve the following problems on recursion


Problem 6 :  
Write a recursive function to calculate the sum of digits of a given positive integer.
sum_of_digits(123) -> 6

Answer 6: '''

def sum_of_digits(n):
    # Base case: If the number is a single digit, return the number itself
    if n < 10:
        return n
    
    # Recursive case: Sum the last digit and recurse with the remaining digits
    return n % 10 + sum_of_digits(n // 10)

    # Example usage
    result = sum_of_digits(123)
    print(result)


'''Problem 7: Fibonacci Series
Write a recursive function to generate the first n numbers of the Fibonacci series.
fibonacci_series(6) -> [0, 1, 1, 2, 3, 5]

Answer 7: '''

def fibonacci_series(n):
    # Base case: If n is 0, return an empty list
    if n == 0:
        return []
    # Base case: If n is 1, return [0]
    elif n == 1:
        return [0]
    # Base case: If n is 2, return [0, 1]
    elif n == 2:
        return [0, 1]
    # Recursive case: Generate the rest of the Fibonacci series
    else:
        fib_series = fibonacci_series(n - 1)
        fib_series.append(fib_series[-1] + fib_series[-2])
        return fib_series

# Example usage
result = fibonacci_series(6)
print(result)

'''Problem 8 : Subset Sum
Given a set of positive integers and a target sum, write a recursive function to determine if there exists a subset
of the integers that adds up to the target sum.
subset_sum([3, 34, 4, 12, 5, 2], 9) -> True'''


def subset_sum(nums, target):
    # Base cases:
    # If the target is 0, an empty subset is a solution
    if target == 0:
        return True
    # If the list is empty, there are no subsets that add up to the target
    if not nums:
        return False

    # Recursive cases:
    # Check two possibilities:
    # 1. Exclude the last element and check if there exists a subset in the rest
    # 2. Include the last element and check if the target can be achieved
    return subset_sum(nums[:-1], target) or subset_sum(nums[:-1], target - nums[-1])

# Example usage
result = subset_sum([3, 34, 4, 12, 5, 2], 9)
print(result)

'''Problem 9: Word Break
Given a non-empty string and a dictionary of words, write a recursive function to determine if the string can be
segmented into a space-separated sequence of dictionary words.
word_break( leetcode , [ leet , code ]) -> True


Answer 9: '''

def word_break(s, word_dict):
    # Base case: If the string is empty, it can be segmented
    if not s:
        return True
    
    # Try to segment the string by checking all prefixes
    for word in word_dict:
        if s.startswith(word):
            # Recur for the remaining substring after removing the matched word
            if word_break(s[len(word):], word_dict):
                return True
    
    # If no segmentation is found, return False
    return False

# Example usage
result = word_break("leetcode", ["leet", "code"])
print(result)


#'Problem 10 : N-Queens
#Implement a recursive function to solve the N Queens problem, where you have to place N queens on an NÃ—N
#chessboard in such a way that no two queens threaten each other.
#n_queens(4)
#'''[
#[".Q..",
#"...Q",
#"Q...",
#"..Q."],
#["..Q.",
#"Q...",
#"...Q",
#".Q.."]
#''']
 

def is_safe(board, row, col, n):
    # Check if there is a queen in the same column
    for i in range(row):
        if board[i][col] == 'Q':
            return False

    # Check upper-left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 'Q':
            return False

    # Check upper-right diagonal
    for i, j in zip(range(row, -1, -1), range(col, n)):
        if board[i][j] == 'Q':
            return False

    return True

def solve_n_queens_util(board, row, n, solutions):
    if row == n:
        # If all queens are placed successfully, add the solution to the list
        solutions.append(["".join(row) for row in board])
        return

    for col in range(n):
        if is_safe(board, row, col, n):
            # Place the queen
            board[row][col] = 'Q'

            # Recur to place the rest of the queens
            solve_n_queens_util(board, row + 1, n, solutions)

            # Backtrack if placing the queen doesn't lead to a solution
            board[row][col] = '.'

def n_queens(n):
    board = [['.' for _ in range(n)] for _ in range(n)]
    solutions = []
    solve_n_queens_util(board, 0, n, solutions)
    return solutions

# Example usage for n=4
n = 4
result = n_queens(n)
for solution in result:
    print(solution)
    
#Below are the outputs to questions 7,8,9,10.
```
